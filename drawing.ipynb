```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int16)
```

```python
blank_img.shape
```

```python
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(100,10),pt2=(400,100),color=(0,255,0),thickness=10)
plt.imshow(blank_img)
```

```python
cv2.rectangle(blank_img,pt1=(200,200),pt2=(300,300),color=(0,0,255),thickness=10)
plt.imshow(blank_img)
```

```python
cv2.circle(blank_img,center=(250,250), radius=100, color= (255,0,0), thickness=10)
plt.imshow(blank_img)
```

```python
cv2.line(blank_img,pt1=(250,0),pt2=(250,512),color=(102,255,255),thickness=10)
cv2.line(blank_img,pt1=(0,250),pt2=(512,250),color=(102,255,255),thickness=10)
plt.imshow(blank_img)
```

```python
font = cv2.FONT_HERSHEY_SIMPLEX
cv2.putText(blank_img,text='Hello',org=(10,425),fontFace=font,fontScale=3,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)
cv2.putText(blank_img,text='World',org=(260,425),fontFace=font,fontScale=3,color=(255,255,255),thickness=3,lineType=cv2.LINE_AA)

plt.imshow(blank_img)
```

```python
blank_img = np.zeros(shape=(512,512,3),dtype=np.int32)
plt.imshow(blank_img)
```

```python
vertices = np.array([[10,400],[100,250],[500,250],[400,400]],dtype=np.int32)
```

```python
vertices.shape
```

```python
pts = vertices.reshape((-1,1,2))
```

```python
pts
```

```python
cv2.polylines(blank_img,[pts],isClosed=True,color=(255,0,0),thickness=5)
plt.imshow(blank_img)
```
